# Customer Churn Analysis - Telco Dataset

In this project, I analyzed customer churn data from a telecommunications company.

## Goal:
- Clean the raw dataset
- Explore the data
- Create visualizations to identify patterns in customer churn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Anett\.ipynb_checkpoints\WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
#basic checks on the dataframe
df.info()
#dataframe size
print("Dataframe size:", df.shape)
#are there any null values?
df.isnull().sum()
```


## Data Cleaning Steps
The following cleaning operations are 
- Converting `TotalCharges` to numeric value and handling invalid values
- Removing rows with missing data
- Converting Yes/No categorical values to binary
- Creating dummy variables for remaining categorical features

```python
#TotalCharges needs to be converted to numeric value
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
#Removing rows with missing data
df.isnull().sum()
df.dropna(inplace=True)
#categorical values to binary
df.replace({'Yes': 1, 'No': 0}, inplace=True)
df.info()
```

```python

df['Churn_Label'] = df['Churn'].map({'No': 'Stayed', 'Yes': 'Churned'})

plt.figure(figsize=(6, 4))
sns.countplot(x='Churn_Label', data=df, palette='pastel')
plt.title('Customer Churn Overview')
plt.xlabel('Customer Status')
plt.ylabel('Number of Customers')
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='gender_Male', hue='Churn', data=df, palette='pastel')

plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.xticks([0, 1], ['Female', 'Male']) #for better readability, 0 and 1 values have been replaced with more descriptive labels (Female, Male)
plt.legend(title='Churn', labels=['Stayed', 'Churned']) 
plt.show()
```

```python
def get_contract_type(row):
    if row['Contract_One year'] == 1:
        return 'One year'
    elif row['Contract_Two year'] == 1:
        return 'Two year'
    else:
        return 'Month-to-month'

df['Contract_Type'] = df.apply(get_contract_type, axis=1)


plt.figure(figsize=(6, 5))
sns.countplot(x='Contract_Type', hue='Churn', data=df, palette='pastel')
plt.title('Customer Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.legend(title='Customer Status', labels=['Stayed', 'Churned'])
plt.xticks(rotation=15)
plt.show()
```

### Insight 
Customers on Month-to-month contracts show the highest churn rate, indicating lower loyalty or satisfaction.
In contrast, customers with One year or Two year contracts tend to stay longer, reflecting the impact of longer-term agreements on retention.
This suggests that incentivizing customers to switch from month-to-month to longer contracts could significantly reduce churn and improve revenue stability.

```python
df['Churn_Label'] = df['Churn'].map({0: 'Stayed', 1: 'Churned'})

plt.figure(figsize=(6, 4))
sns.boxplot(x='Churn_Label', y='MonthlyCharges', data=df, palette='pastel') 

plt.title('Monthly Charges by Customer Status')
plt.xlabel('Customer Status')
plt.ylabel('Monthly Charges (USD)')
plt.show()
```

### Insight

The boxplot reveals that customers who churn tend to have higher monthly charges compared to those who stay.  
This suggests that pricing could be a significant factor influencing customer churn, highlighting the need for targeted retention strategies such as tailored pricing plans or discounts for high-paying customers to improve loyalty and reduce churn rates.

```python
plt.figure(figsize=(12, 10))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, cmap='coolwarm', annot=False, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Insight

- **Tenure vs. Churn:** Longer-tenure customers are much less likely to churn.
- **Monthly Charges vs. Churn:** Higher monthly fees are slightly associated with increased churn.
- **Service Usage:** Users of Online Security and Tech Support services tend to stay longer.
- **Contract Type:** Customers with two-year contracts have the lowest churn rates.
